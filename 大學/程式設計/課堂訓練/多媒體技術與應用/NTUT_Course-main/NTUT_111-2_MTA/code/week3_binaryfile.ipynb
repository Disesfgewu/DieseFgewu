# 附錄: 圖片轉成二進位數據儲存

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img_path = 'Tommy.jpg'
img_bgr = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
print(img_bgr.shape)
plt.subplot(1,2,1)
plt.imshow(img_bgr)
plt.title('cv2 BGR')
plt.subplot(1,2,2)
plt.imshow(img_rgb)
plt.title('cv2 RGB')
plt.show()
```

```python
# # https://kuanyuchen.gitbooks.io/python3-tutorial/content/er_jin_zhi_chu_li_fang_shi.html
print(img_rgb.shape)
with open('Tommy_rgb.bin', 'wb') as f:
    f.write(bytes(img_rgb))
```

從下範例可以發現二進制儲存會以一個很大的array儲存

```python
with open('Tommy_rgb.bin', 'rb') as f:
    data = f.read()
c_img = np.array(bytearray(data))
print(c_img.shape)
```

所以讀入後需要reshape為原始圖片大小

```python
c_img = c_img.reshape(480, 640, 3)
plt.imshow(c_img)
plt.show()
```
