<font size=3>
    假設
$f(x)=x^4-50x^3-x+1$<br>
$$
x^* = \min_x{f(x)}
$$
    
請求$x^*$

```python
import numpy as np
def f(x):
    return x**4 -50*x**3-x+1
def g(x):
    return 4*x**3-150*x**2-1

x_hat=np.array([30])
lr=0.0004
for i in range(100):
    x_hat = x_hat - lr * g(x_hat)

print('x_star={}'.format(x_hat))
print('f(x_star)={}'.format(f(x_hat)))
```

<font size=3>
    假設
$f(x)=x_2^2+x_1-10$<br>
$$
x^* = \min_x{f(x)}
$$
    
請求$x^*$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x1,x2):
    return x2**2 + x1 - 10
def g(x1,x2):
    return np.array([1, 
                     2*x2])

x_hat=np.array([5, 5])
plt.plot(x_hat[0], x_hat[1],'b*')
lr=0.1
for i in range(50):
    x_hat = x_hat - lr * g(x_hat[0], x_hat[1])
    plt.plot(x_hat[0], x_hat[1],'k*')

plt.plot(x_hat[0], x_hat[1],'r*')
plt.xlim([-6,6])
plt.ylim([-6,6])
print('x_star={}'.format(x_hat))
print('f(x_star)={}'.format(f(x_hat[0],x_hat[1])))
```
