# 特徵萃取法(Feature Extraction) - PCA

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder 
def load_classification_example():
    filename='.\data_weight_multlfeatutes.xlsx'
    df = pd.read_excel(filename)
    labelencoder = LabelEncoder()
    df["性別"] = labelencoder.fit_transform(df["性別"]) # 將"性別"裡的字串自動轉換成數值
    df["手機品牌"] = labelencoder.fit_transform(df["手機品牌"]) # 將"手機品牌"裡的字串自動轉換成數值
    X = np.array(df[['身高', '體重', '手機品牌','體脂肪']])
    Y = np.array(df['性別'])
    return X, Y
````

### PCA + 分類器(LogisticRegression)

````python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

X, Y = load_classification_example()
clf = LogisticRegression()
clf.fit(X,Y)
print('All features: accuracy:{}'.format(clf.score(X, Y)))

pca = PCA(n_components=4)
pca.fit(X)
X_ = pca.transform(X)
for k in [1,2,3,4]:
    clf = LogisticRegression()
    clf.fit(X_[:,:k],Y)
    print('PCA(pc={})): accuracy:{}'.format(k, clf.score(X_[:,:k], Y)))
````

### PCA + 分類器(QuadraticDiscriminantAnalysis)

````python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis  import QuadraticDiscriminantAnalysis

X, Y = load_classification_example()
clf = QuadraticDiscriminantAnalysis()
clf.fit(X,Y)
print('All features: accuracy:{}'.format(clf.score(X, Y)))

pca = PCA(n_components=4)
pca.fit(X)
X_ = pca.transform(X)
for k in [1,2,3,4]:
    clf = QuadraticDiscriminantAnalysis()
    clf.fit(X_[:,:k],Y)
    print('PCA(pc={})): accuracy:{}'.format(k, clf.score(X_[:,:k], Y)))
````

### 利用sklearn內建的資料庫來測試速度
digits在Sklean是一個1797張圖片的數字資料庫。<br>
我們將利用它來測試不同特徵選取的速度。<br>
並且我們也可以看主成分變異累積貢獻百分比來分析<br>

````python
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits

digits = load_digits()
X = digits.data
Y = digits.target

pca = PCA(n_components=64)
pca.fit(X)
cum_pca=0
for i, tmp in enumerate(pca.explained_variance_ratio_):
    cum_pca+=tmp
    print('{}個主成分變異貢獻百分比:{}'.format(i+1, cum_pca*100))
    
````

從上結果可以
1. 發現累積到13個主成分後，可以保留原資料80%的變異。<br>
2. 累積到21個主成分後，可以保留原資料90%的變異。<br>
我們進一步分析不同主成分取出後計算QDA的分類正確率。

````python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis  import QuadraticDiscriminantAnalysis
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import time 
import warnings
import numpy as np
warnings.filterwarnings("ignore")

digits = load_digits()
X = digits.data
Y = digits.target


clf = QuadraticDiscriminantAnalysis()
clf.fit(X,Y)
print('All features: accuracy:{}'.format(clf.score(X, Y)))

st = time.time()
pca = PCA(n_components=64)
pca.fit(X)
X_ = pca.transform(X)
acc=[]
for k in range(64):
    k+=1
    clf = QuadraticDiscriminantAnalysis()
    clf.fit(X_[:,:k],Y)
    print('PCA(pc={})): accuracy:{}'.format(k, clf.score(X_[:,:k], Y)))
    acc.append(clf.score(X_[:,:k], Y))
acc=np.array(acc)
print('{} of PCs can get best performance with accuracy: {}'.format(np.argmax(acc)+1,
                                                                    acc[np.argmax(acc)]
                                                                   ))
print('whole time:{:5f}s'.format(time.time()-st))
````
