# 利用SVD的技術來壓縮影像

利用singular value decomposition (SVD)的分解技術，分解給定的影像後，刪除數值較小的數個singular value，再組成其近似影像，藉此降低影像的儲存空間。
給定一張影像$A \in R^{m \times n}$，假定其SVD的結果為
$$
A=U \Sigma V^T
$$
$U \in R^{m \times m}$ : orthogonal matrix <br>
$V \in R^{n \times n}$ : orthogonal matrix <br>
$\Sigma \in R^{m \times n}$ : matrix <br>

$$
\Sigma = \left[
    \begin{array}{ccc|c}
      \sigma_0 & & 0 & \\
       & \ddots&  & 0\\
      0 & & \sigma_p 
    \end{array}
\right] ,   {\rm if }  m<n
$$  

$$
\Sigma = \left[
    \begin{array}{ccc}
    \sigma_0  &        & 0 \\
              & \ddots &   \\
      0       &        & \sigma_p \\
    \hline
    &0 &
    \end{array}
\right]  ,    {\rm if }   m \geq n 
% 內層第二行表格組結束
$$

$\sigma_1 \geq \sigma_2 \geq \cdots  \geq \sigma_p \geq 0, p=min\{m,n\}$

假設 <br>
$U = [u_1, u_2, \cdots, u_m]$ and $V=[v_1,v_2,\cdots,v_n]$ <br>

所以 <br>
$$A = \sigma_1 u_1 v_1^T+\sigma_2 u_2 v_2^T+\cdots+\sigma_p u_p v_p^T$$

**壓縮方式**:<br>
我們可以取前面最大的t個singular value計算出壓縮影像 $\tilde{A}$來近似影像$A$，即
$$A = \sigma_1 u_1 v_1^T+\sigma_2 u_2 v_2^T+\cdots+\sigma_t u_t v_t^T, t\leq p$$
 


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def TakeSVD(A,p):
    A=np.array(A, dtype=np.double)
    U,S,V = np.linalg.svd(A, full_matrices=False)
    print('U:{}, S:{}, V:{}'.format(U.shape, S.shape, V.shape))
    S=np.diag(S)
    rdU=U[:,0:p]
    rdS=S[0:p,0:p]
    rdV=V[0:p, :]
    print('rdU:{}, rdS:{}, rdV:{}'.format(rdU.shape, rdS.shape, rdV.shape))

    rdB = np.dot(rdU, np.dot(rdS, rdV))
    return rdB



img_path = 'example.png'
img_bgr = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
print('origianl image shape:{}'.format(img_rgb.shape))
plt.imshow(img_rgb)
plt.show()

for t in [50, 200, 400]:
    print('-------')
    print('dimension:{}'.format(t))
    img_rgb_hat = img_rgb.copy()
    for i in range(3):
        img_rgb_hat[:,:,i]=TakeSVD(img_rgb_hat[:,:,i],t);
    img_rgb_hat = np.uint8(img_rgb_hat)
    MSE=np.mean((img_rgb_hat-img_rgb)^2)
    print('MSE:{}'.format(MSE))
    plt.imshow(img_rgb_hat)
    plt.show()
    
```

# 範例解說:<br>
$ A \in R^{665 \times 900 \times 3 } $
$$ A=U \Sigma V^T 
$$
$U \in R^{665 \times 665}$ : orthogonal matrix <br>
$V \in R^{900 \times 900}$ : orthogonal matrix <br>
$\Sigma \in R^{665 \times 900}$ :  matrix <br>
因為numpy的SVD計算$\Sigma$只會保留非0數字，所以Singular value會剩下
$p=min\{665,900\}=665$ <br> 
$\sigma_1 \geq \sigma_2 \geq \cdots  \geq \sigma_{665} $

## 原始儲存大小: 整張圖儲存
$ A \in R^{665 \times 900 \times 3 } $
$$ 665 \times 900 \times 3 = 1,795,500 $$
1,795,500 byte = 1753.41796875 KB =  1.7123222351074219MB

## 如果利用SVD儲存: 儲存大小 
$$U \in R^{665 \times 665} $$
$$S \in R^{665}$$
$$V \in R^{900 \times 900}$$
$$ 665 \times 665 + 665 + 900\times 900 = 1,252,890 $$
1,252,890 byte = 1223.525390625 KB =1.1948490142822266 MB


## 如果利用SVD儲存: 假設t設定為400
我們只需要儲存
$$rdU \in R^{665 \times 400} $$
$$rdS \in R^{400}$$
$$rdV \in R^{400 \times 900}$$

$$ 665 \times 400 + 400 + 400\times 900 = 626,400 $$
626,400 byte = 611.71875 KB

壓縮比= 626,400/1,795,500 = 0.34887 <br>
壓縮比= 626,400/1,252,890 = 0.49996 <br>
