## Q1 多元回歸(Multiple regression) (25%)


$$
({\bf{x}}_i,y_i) , {\bf{x}}_i = \begin{bmatrix} 1 \\ x_{i1} \\ \vdots \\ x_{id}\\ \end{bmatrix} \in R^{d+1}, \forall i=1,2,...,n 
$$

$$
{\bf{Y}} = \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_n\\ \end{bmatrix}, 
{\bf{\beta}} = \begin{bmatrix} \beta_0 \\ \beta_1 \\ \vdots \\ \beta_n\\ \end{bmatrix},
{\bf{X}} = \begin{bmatrix} \bf{X}_1^T \\ \bf{X}_2^T  \\ \vdots \\ \bf{X}_n^T \\ \end{bmatrix}
 = \begin{bmatrix} 1&x_{11}&...&x_{1d} \\ \ 1&x_{21}&...&x_{2d}  \\ \vdots \\ \ 1&x_{n1}&...&x_{nd} \\ \end{bmatrix} 
$$

**Regression:** $$\bf{Y}=  {\bf{X}}{\bf{\beta}}$$

$$
loss({\bf{\beta}})=(\bf{Y}-\hat{\bf{Y}})^T(\bf{Y}-\hat{\bf{Y}})
$$

<font size=4 color=red>**找${\bf{\beta}}$的closed-form solution**

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder 
import matplotlib.pyplot as plt

def load_classification_example():
    filename='.\data_weight_multlfeatutes.xlsx'
    df = pd.read_excel(filename)
    labelencoder = LabelEncoder()
    df["性別"] = labelencoder.fit_transform(df["性別"]) # 將"性別"裡的字串自動轉換成數值
    df["手機品牌"] = labelencoder.fit_transform(df["手機品牌"]) # 將"手機品牌"裡的字串自動轉換成數值
    X = np.array(df[['身高', '體重', '手機品牌','體脂肪']])
    Y = np.array(df['性別'])
    return X, Y
data, target = load_classification_example()

data[:,0]=data[:,0]/200
data[:,1]=data[:,1]/100
data=data[:,0:2]

pos0=np.where(target==0)[0]
pos1=np.where(target==1)[0]
plt.figure()
plt.plot(data[pos0,0],data[pos0,1],'r*')
plt.plot(data[pos1,0],data[pos1,1],'b*')
plt.show()
````

````python
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torch
torch.manual_seed(202404)
data = torch.tensor(data, dtype=torch.float).requires_grad_(False)
target = torch.tensor(target, dtype=torch.float).requires_grad_(False)
````

````python
use_cuda=0
device = torch.device("cuda" if (torch.cuda.is_available() & use_cuda) else "cpu")
# define NN architecture
class SLP(nn.Module):
    def __init__(self):
        super(SLP,self).__init__()
        self.linear = torch.nn.Linear(2, 1)
    def forward(self,x):
        y_pred = torch.sigmoid(self.linear(x))
        return y_pred
model1 = SLP().to(device)
loss = torch.nn.BCELoss()
optimizer = optim.SGD(model1.parameters(), lr=0.1, momentum=0.9)
log_loss_exp1 = []
model1.train()
data, target = data.to(device), target.to(device)
for epoch in range(10000):
    output_slp = model1(data)
    loss_slp = loss(output_slp.squeeze(), target)
    model1.zero_grad()
    loss_slp.backward()
    optimizer.step()
    log_loss_exp1.append(loss_slp.item())
````

````python
model2 = SLP().to(device)
loss = torch.nn.BCELoss()
optimizer = optim.SGD(model2.parameters(), lr=0.01, momentum=0.9)
log_loss_exp2 = []
model2.train()
data, target = data.to(device), target.to(device)
for epoch in range(10000):
    output_slp = model2(data)
    loss_slp = loss(output_slp.squeeze(), target)
    model2.zero_grad()
    loss_slp.backward()
    optimizer.step()
    log_loss_exp2.append(loss_slp.item())
````

````python
print('model1: lr=0.1')
print(model1.linear.weight)
print(model1.linear.bias)

print('model2: lr=0.01')
print(model2.linear.weight)
print(model2.linear.bias)
````

````python
plt.figure()
plt.plot(log_loss_exp1,label='lr=.1')
plt.plot(log_loss_exp2,'r',label='lr=.01')
plt.legend()
````

````python
X1 = np.linspace(.7,1.0,200)
X2 = np.linspace(0.4,1.0,200)
R1 = np.zeros((200,200))
R2 = np.zeros((200,200))
model1.eval()
model2.eval()
for i,x1 in enumerate(X1):
    for j,x2 in enumerate(X2):
        x = torch.tensor([x1,x2], dtype=torch.float).requires_grad_(False)
        output1 = model1(x)
        output2 = model2(x)
        R1[i,j]=output1[0].detach().numpy()
        R2[i,j]=output2[0].detach().numpy()
````

````python
XX1, XX2 = np.meshgrid(X1, X2)
fig, ax = plt.subplots()
c = ax.pcolormesh(XX1, XX2, R1, cmap='RdBu', vmin=0, vmax=1)
ax.set_title('lr=0.1')
fig.colorbar(c, ax=ax)
pos0=np.where(target==0)[0]
pos1=np.where(target==1)[0]
ax.plot(data[pos0,0],data[pos0,1],'r*')
ax.plot(data[pos1,0],data[pos1,1],'b*')
plt.show()

XX1, XX2 = np.meshgrid(X1, X2)
fig, ax = plt.subplots()
c = ax.pcolormesh(XX1, XX2, R2, cmap='RdBu', vmin=0, vmax=1)
ax.set_title('lr=0.01')
fig.colorbar(c, ax=ax)
pos0=np.where(target==0)[0]
pos1=np.where(target==1)[0]
ax.plot(data[pos0,0],data[pos0,1],'r*')
ax.plot(data[pos1,0],data[pos1,1],'b*')
plt.show()
````
